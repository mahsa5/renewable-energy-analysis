# Global Renewable Energy Production Analysis (1990-2017)

## Project Overview
This analysis examines the trends in global renewable energy production across four major sources: Hydroelectric, Biofuel, Solar PV, and Geothermal.

## Key Findings

### 1. Hydroelectric Power Dominance
- Remains the largest renewable energy source
- Grew from 2,192 TWh (1990) to 4,197 TWh (2017)
- Accounts for 71% of total renewable energy in 2017

### 2. Explosive Growth in Biofuel
- Increased from 4 TWh to 1,127 TWh (280x growth!)
- Second largest source by 2017 (19% share)
- Most dramatic growth trajectory

### 3. Solar Revolution
- Started near zero (0.09 TWh in 1990)
- Reached 444 TWh by 2017
- Represents the fastest-growing renewable technology
- Major acceleration after 2010

### 4. Steady Geothermal
- Modest growth from 36 TWh to 85 TWh
- Smallest contributor (1.4% in 2017)
- Most stable growth pattern

## Environmental Implications
The rapid expansion of solar and biofuel production after 2010 indicates a global shift toward diversified renewable energy portfolios, reducing dependence on traditional fossil fuels.

## Data Source
Kaggle: Renewable Power Generation Dataset

```python
# محاسبه درصد رشد هر انرژی
growth_rates = {
    'Hydro': ((last_year['Hydro(TWh)'] - first_year['Hydro(TWh)']) / first_year['Hydro(TWh)']) * 100,
    'Biofuel': ((last_year['Biofuel(TWh)'] - first_year['Biofuel(TWh)']) / first_year['Biofuel(TWh)']) * 100,
    'Solar': ((last_year['Solar PV (TWh)'] - first_year['Solar PV (TWh)']) / first_year['Solar PV (TWh)']) * 100,
    'Geothermal': ((last_year['Geothermal (TWh)'] - first_year['Geothermal (TWh)']) / first_year['Geothermal (TWh)']) * 100
}

# نمایش نتایج
print("Growth Rates (1990-2017):\n")
for energy, rate in growth_rates.items():
    print(f"{energy:15} {rate:>10.1f}%")

# رسم نمودار
plt.figure(figsize=(10, 6))
colors_bar = ['#3498db', '#2ecc71', '#f39c12', '#e74c3c']
bars = plt.bar(growth_rates.keys(), growth_rates.values(), color=colors_bar, edgecolor='black', linewidth=1.5)

plt.title('Growth Rate of Renewable Energy Sources (1990-2017)', fontsize=14, fontweight='bold')
plt.ylabel('Growth Rate (%)', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# اضافه کردن مقادیر روی میله‌ها
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}%',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n✅ Growth analysis completed!")
```

```python
# مقایسه 1990 vs 2017 - Bar Chart
years_compare = ['1990', '2017']
hydro_vals = [first_year['Hydro(TWh)'], last_year['Hydro(TWh)']]
biofuel_vals = [first_year['Biofuel(TWh)'], last_year['Biofuel(TWh)']]
solar_vals = [first_year['Solar PV (TWh)'], last_year['Solar PV (TWh)']]
geo_vals = [first_year['Geothermal (TWh)'], last_year['Geothermal (TWh)']]

x = range(len(years_compare))
width = 0.2

fig, ax = plt.subplots(figsize=(12, 7))

bar1 = ax.bar([i - 1.5*width for i in x], hydro_vals, width, label='Hydro', color='#3498db')
bar2 = ax.bar([i - 0.5*width for i in x], biofuel_vals, width, label='Biofuel', color='#2ecc71')
bar3 = ax.bar([i + 0.5*width for i in x], solar_vals, width, label='Solar', color='#f39c12')
bar4 = ax.bar([i + 1.5*width for i in x], geo_vals, width, label='Geothermal', color='#e74c3c')

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Energy Production (TWh)', fontsize=12)
ax.set_title('Renewable Energy Production: 1990 vs 2017 Comparison', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(years_compare)
ax.legend(fontsize=10)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ Comparison chart created!")
```

```python
# Stacked Area Chart - نشون میده سهم هر انرژی در طول زمان
plt.figure(figsize=(14, 7))

plt.fill_between(df['Year'], 0, df['Hydro(TWh)'],
                 label='Hydro', alpha=0.7, color='#3498db')
plt.fill_between(df['Year'], df['Hydro(TWh)'],
                 df['Hydro(TWh)'] + df['Biofuel(TWh)'],
                 label='Biofuel', alpha=0.7, color='#2ecc71')
plt.fill_between(df['Year'], df['Hydro(TWh)'] + df['Biofuel(TWh)'],
                 df['Hydro(TWh)'] + df['Biofuel(TWh)'] + df['Solar PV (TWh)'],
                 label='Solar', alpha=0.7, color='#f39c12')
plt.fill_between(df['Year'], df['Hydro(TWh)'] + df['Biofuel(TWh)'] + df['Solar PV (TWh)'],
                 df['Hydro(TWh)'] + df['Biofuel(TWh)'] + df['Solar PV (TWh)'] + df['Geothermal (TWh)'],
                 label='Geothermal', alpha=0.7, color='#e74c3c')

plt.title('Total Renewable Energy Production Over Time (Stacked)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Cumulative Energy Production (TWh)', fontsize=12)
plt.legend(loc='upper left', fontsize=11)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Stacked area chart created!")
```

```python
# محاسبه مجموع کل انرژی تجدیدپذیر در هر سال
df['Total Renewable'] = df['Hydro(TWh)'] + df['Biofuel(TWh)'] + df['Solar PV (TWh)'] + df['Geothermal (TWh)']

# نمودار رشد کل
plt.figure(figsize=(14, 7))

plt.plot(df['Year'], df['Total Renewable'], marker='o', linewidth=3,
         color='#27ae60', markersize=6, label='Total Renewable Energy')

# اضافه کردن خط روند
z = np.polyfit(df['Year'], df['Total Renewable'], 1)
p = np.poly1d(z)
plt.plot(df['Year'], p(df['Year']), "--", color='red', linewidth=2, alpha=0.7, label='Trend Line')

plt.title('Total Global Renewable Energy Production (1990-2017)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Production (TWh)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(alpha=0.3)

# اضافه کردن annotations برای نقاط کلیدی
total_1990 = df[df['Year'] == 1990]['Total Renewable'].values[0]
total_2017 = df[df['Year'] == 2017]['Total Renewable'].values[0]

plt.annotate(f'{total_1990:.0f} TWh', xy=(1990, total_1990),
             xytext=(1992, total_1990 + 300),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10, fontweight='bold')

plt.annotate(f'{total_2017:.0f} TWh', xy=(2017, total_2017),
             xytext=(2014, total_2017 + 300),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n📊 Key Statistics:")
print(f"Total Renewable Energy in 1990: {total_1990:.0f} TWh")
print(f"Total Renewable Energy in 2017: {total_2017:.0f} TWh")
print(f"Overall Growth: {((total_2017 - total_1990) / total_1990 * 100):.1f}%")
print("\n✅ Final visualization completed!")
```
